```python
# Import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

import joblib

# Load dataset
df = pd.read_csv("animal_disease_dataset22.csv")
df.head()
```

```python
# Cek data
print(df.info())

print(df.describe())

print(df.isnull().sum())

print(df.duplicated().sum())

print(df['Disease'].value_counts())
```

```python
# Copy dataframe
df_corr = df.copy()

# Label Encoding semua kolom kategorikal
le = LabelEncoder()

for col in df_corr.columns:
    if df_corr[col].dtype == 'object':
        df_corr[col] = le.fit_transform(df_corr[col])

# Hitung matriks korelasi
corr_matrix = df_corr.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Heatmap Korelasi")
plt.show()
```

```python
# Target kolom
target_column = 'Disease'

# Pisahkan fitur & target
X = df.drop(target_column, axis=1)
y = df[target_column]

# One-Hot Encoding untuk fitur kategorikal
X_encoded = pd.get_dummies(X)

# Label Encoding untuk target
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("\nPra-pemrosesan selesai. Contoh data fitur setelah di-encode:")
display(X_encoded.head())
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"Ukuran data latih: {X_train.shape}")
print(f"Ukuran data uji: {X_test.shape}")
```

```python
# Train model
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)
```

```python
# Prediksi
y_pred = model.predict(X_test)
```

```python
# Evaluasi
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

print("Akurasi:", accuracy_score(y_test, y_pred))
```

```python
# Visualisasi Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Distribusi target
plt.figure(figsize=(8, 4))
sns.countplot(x=y, order=y.value_counts().index)
plt.title("Distribusi Kelas Target (Penyakit)")
plt.xlabel("Penyakit")
plt.ylabel("Jumlah")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=df['Animal'], order=df['Animal'].value_counts().index)
plt.title("Distribusi Kolom Animal")
plt.xlabel("Jenis Hewan")
plt.ylabel("Jumlah")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Buat tabel pivot: jumlah hewan per penyakit
pivot_table = pd.crosstab(df['Animal'], df['Disease'])

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="Blues")
plt.title("Heatmap Frekuensi Animal vs Disease")
plt.xlabel("Disease")
plt.ylabel("Animal")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import joblib

joblib.dump(model, "random_forest_model.pkl")
joblib.dump(le, "label_encoder.pkl")
joblib.dump(X_encoded.columns.tolist(), 'feature_columns.pkl')
```
